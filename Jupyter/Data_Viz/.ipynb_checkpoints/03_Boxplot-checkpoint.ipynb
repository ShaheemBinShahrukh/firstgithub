```python
# Importing the required module
import seaborn as sns

# Use to set style of background of plot
sns.set(style="whitegrid")

# Loading data-set
tip = sns.load_dataset("tips")

sns.boxplot (x= "tip", y= "day", hue= "smoker", data= tip,
            palette= "Set2", dodge=True. orient= "")
```

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

kashti = sns.load_dataset ("titanic")
sns.boxplot (x= "survived", y= "age", data=kashti)
```

```python
p1 = sns.boxplot (x= "survived", y= "age", showmeans =True, data=kashti)
```

```python
sns.boxplot(x="survived", y="age", showmeans=True,
            meanprops={"marker": "+", "markersize": 12,
                       "markeredgecolor": "red"}, data=kashti)
# Show Labels
plt.xlabel("How Many Survived", size = 8, weight= 'bold')
plt.ylabel ("Age (years)", size = 10, weight= 'bold'),
plt.title ("Box Plot of kitany doobay aur kitnay bach gaye", size= 14, weight= 'bold')
plt.show()
```
